```python
import pandas as pd
```

```python
forest = pd.read_csv('forest.csv', engine ='python', encoding='EUCKR')
```

```python
# forest = forest.loc[forest['경도'].isnull()].dropna()
# forest[['위도', '경도']].dropna()
forest = forest.dropna(subset=["위도"])
```

```python
sel = ['위도','경도']
temp_forest = forest[sel]
print(temp_forest)
```

```python
import folium
from folium import plugins
import numpy as np 
import os
```

```python
num = temp_forest.to_numpy()
# len_num =len(num)
print(num)
# print(len(num))
```
